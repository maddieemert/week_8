1. What makes a model "linear"? "Linear" in what?

The parameters/coefficients are what make a model linear. Each parameter must appear as a constant multiplier, and the model must be able to be written as a weighted sum of terms.

2. How do you interpret the coefficient for a dummy/one-hot-encoded variable? (This is a trick question, and the trick involves how you handle the intercept of the model.) There's further explanation at the end of this document, if needed.

The coefficient for a dummy variable represents how much that category differs from the baseline category, and the intercept is what defines that baseline.

3. Can linear regression be used for classification? Explain why, or why not.

Linear regression can be used for classification as long as the classes are encoded as numbers (0 and 1), which will then output a continuous value that can be used for prediction. However, this is somewhat of a forced method, as its assumptions and outputs don't align with the goals of classification. Logistic regression would serve as a better alternative because it constrains outputs between 0 and 1 and uses a loss function better suited for classification.

4. If you have a high accuracy on training but low on testing, what might be the problem? It also might be you see a pattern in the residuals. 

This problem reflects overfitting, as high training accuracy means the model fits the training data well but low testing accuracy means the model doesn't generalize well to new data. Residuals should also look random if the model is appropriate, so a clear pattern means the model is missing structure in the data.

One way to incorporate nonlinear relationships between x and y is to use a model that is inherently nonlinear, such as decision trees or support vector machines with nonlinear kernels. This ensures that complexity is handled directly, as these models don't assume a straight-line relationship. Another way to incorporate nonlinear relationships between x and y is to transform input features into nonlinear versions of themselves and then apply linear regression. You can use polynomial expansion then fit a linear model on the new features, ensuring that the model is linear in coefficients yet nonlinear overall due to feature transformations. 

6. What is the interpretation of the intercept? A slope coefficient for a variable? The coefficient for a dummy/one-hot-encoded variable?

The intercept represents the predicted value of y when the input variables are equal to 0, serving as the baseline value. A slope coefficient for a numerical variable represents the change in y for a one-unit increase in that variable, with the assumption that all other variables are held constant. The coefficient for a dummy/one-hot-encoded variable represents the difference in y between that category and the baseline category (how much higher or lower it is).
